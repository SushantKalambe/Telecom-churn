```python
def name(data,column1,column2,column3,column4,col_name,value1,value2):
    data['avg_1']= (data[column1]+data[column2])/2
    data['avg_2']= (data[column3]+data[column4])/2
    
    data[col_name]=np.where(((data['avg_1'] > data['avg_2'])),value1,value2)
    net=data[col_name].value_counts()
    val_1=(data[data[col_name] == net.index[0]])['label'].value_counts()
    val_2=(data[data[col_name] == net.index[1]])['label'].value_counts()
    vals=pd.concat([pd.DataFrame(val_1),pd.DataFrame(val_2.to_frame())], axis=1, sort=False).T
    frame=pd.concat([net.reset_index(),vals.reset_index()],axis=1,sort=False)
    val_1_rate=100*(val_1[1]/2589)#net[0]
    val_2_rate=100*(val_2[1]/2589)#net[1]
    return  sns.countplot(x=col_name ,hue='label',data=data),print(val_1_rate),print(val_2_rate),print(frame)
           
```

```python
def name_x(data,column1,column2,column3,column4,column5,column6,col_name,value1,value2,value3):
    data['avg_1']= (data[column1]+data[column2])/2
    data['avg_2']= (data[column3]+data[column4])/2
    data['avg_3']= (data[column5]+data[column6])/2
    
   
    data[col_name]=np.where((data['avg_1']>data['avg_2']) & (data['avg_1']>data['avg_3']), value1, 
         (np.where((data['avg_2']>data['avg_1']) & (data['avg_2']>data['avg_3']), value2, value3)))
    
    net=data[col_name].value_counts()
    val_1=(data[data[col_name] == net.index[0]])['label'].value_counts()
    val_2=(data[data[col_name] == net.index[1]])['label'].value_counts()
    val_3=(data[data[col_name] == net.index[2]])['label'].value_counts()
    vals=pd.concat([pd.DataFrame(val_1),pd.DataFrame(val_2.to_frame()),pd.DataFrame(val_3.to_frame())], axis=1, sort=False).T
    frame=pd.concat([net.reset_index(),vals.reset_index()],axis=1,sort=False)
    val_1_rate=100*(val_1[1]/2589)#net[0]
    val_2_rate=100*(val_2[1]/2589)#net[1]
    val_3_rate=100*(val_3[1]/2589)
    return  sns.countplot(x=col_name ,hue='label',data=data),print(val_1_rate),print(val_2_rate),print(val_3_rate),print(frame)
```

```python
main['total_call_mou_6']=(main['total_ic_mou_6']+main['total_og_mou_6'])
main['total_call_mou_7']=(main['total_ic_mou_7']+main['total_og_mou_7'])
main['total_avg_call']=(main['total_call_mou_6']+main['total_call_mou_7'])/2
main['total_data_6']=(main['vol_2g_mb_6']+main['vol_3g_mb_6'])
main['total_data_7']=(main['vol_2g_mb_7']+main['vol_3g_mb_7'])
main['total_avg_data']=(main['total_data_6']+main['total_data_7'])/2
main['call']=np.where(((main['total_avg_call'] > main['total_avg_call'].mean())),'high_call','low_call')
main['data']=np.where(((main['total_avg_data'] > main['total_avg_data'].mean())),'high_data','low_data')
main['call_data']=main['call'] + '_' + main['data']
```

###  The other segment (2nd high) which contains 31% users are high calling users also have more churn rate.
